### Remove extra text

```python
# df['column1'] = df['column1'].str.replace('extra text:', '')

# Remove Writtenby: from the author column
df['author'] = df['author'].str.replace('Writtenby:', '')

# Remove Narratedby: from the narrator column

df['narrator'] = df['narrator'].str.replace('Narratedby:', '')

# Display data
df
```

### Extract data & split into columns

```python
# .sample(n=n)

# Display first 20 data from stars column
df.stars.sample(n=20)
```

The results are combined with stars and ratings as well as other raw audible data is not rated yet.

```python
# df['new_column1'] = df['new_column2'].str.extract


# Extract number of stars into new_stars and convert into float
df['new_stars'] = df['stars'].str.extract('^([\d.]+)').astype(float)

# df['new_ratings'] = df['stars'].str.replace(',', '').str.extract('(\d+) rating').astype(float)

# Replace the comma, extract number of ratings into new_ratings and convert into float
df['new_ratings'] = df['stars'].str.replace(',', '').str.extract('(\d+) rating').astype(float)

# df[['new_column1', 'new_column2']]

# Display the new_stars and new_ratings columns
df[['new_stars', 'new_ratings']]
```

### Check for duplicates & null values again after extract & split into new columns

```python
# Check for duplicates
df.duplicated().sum()
```

```python
# Check number of rows where particular columns of null values
# df.isna().sum()

df.isnull().sum()
```

There are 72417 (not rated yet) null values (NaN) found in new columns after extract & split.

### Replace nan with 0

```python
# df['new_column1']=df['new_column1'].fillna('new_column1'),mode()[0])
# .mode() means values that appears most often

df['new_stars']= df['new_stars'].fillna(df['new_stars'].mode()[0])
df['new_ratings']= df['new_ratings'].fillna(df['new_ratings'].mode()[0])
```

```python
# Check nan after splitting into new_stars & new_ratings columns
df.isna().sum()
```

```python
# Display data after extracting & splitting into new columns
df
```

### Drop unused column

```python
# df.drop(columns=['old_column'], axis=1, inplace=True)
# Inplace means reassign values to df_old or df_new after dropping columns
# axis=1 y axis

# Drop old stars column
df.drop(columns=['stars'], axis=1, inplace=True)

# Display data after dropping stars column
df
```

### Change data types

- Convert `price` to float

- Convert `new_stars` to category

- Convert `releasedate` to datetime

```python
# .sample(n=n)

# Display random 20 data from price column
df.price.sample(n=20)
```

There is a need to remove comma and free before converting the data.

```python
# Replace the comma with ''
df['price'] = df.price.str.replace(',', '')

# Replace 'Free' with 0
df['price'] = df.price.str.replace('Free', '0')

# Convert price to float
df['price'] = df.price.astype(float)

# Display data
df
```

```python
# df.new_column.unique()
# .unique() means find unique values from a series

# Convert new_stars column into category due to take into small discrete values
df.new_stars.unique()
```

```python
# df['new_column'] = df.new_column.astype('category')
# .astype means change data type

# Convert new_stars to category
df['new_stars'] = df.new_stars.astype('category')
```

```python
# to_datetime() means convert the dates into datetime objects

# Convert releasedate to datetime
df['releasedate'] = pd.to_datetime(df.releasedate)

# Print full summary & check data types again
df.info()
```

### Extract data & create new column

old
time column combines hours & minutes

new_time column 
- hr, hrs -> consolidate as hr
- min, mins -> consolidate as min
- Less than 1 minute -> round to 1 min

```python
# .sample(n=n)

# Display random 20 data from time column
df.time.sample(n=20)
```

```python
# Search the entries in the time column for different spellings of min, mins or minutes
df.time[df.time.str.contains('minute')].sample(n=10)
```

```python
# df['old_column'] = df.time.str.replace('hrs', 'hr')

# Replace hrs, mins, and 'Less than 1 minute'
df['time'] = df.time.str.replace('hrs', 'hr')
df['time'] = df.time.str.replace('mins', 'min')
df['time'] = df.time.str.replace('Less than 1 minute', '1 min')
```

```python
# df[['column']]

# Check time column after replacing the time format
df[['time']]
```

Extract the number of hours & minutes from time column, and combine into a new_time column.

```python
# Extract hours and convert to integer
hrs = df['time'].str.extract('(\d+) hr').fillna(0).astype(int)

# Extract minutes and convert to integer
mins = df['time'].str.extract('(\d+) min').fillna(0).astype(int)

# hr60-min60-sec60
# Combine hours and minutes into the new_time column
df['new_time'] = hrs * 60 + mins

df[['new_time']]
```

```python
# df.drop(columns=['old_column'], axis=1, inplace=True)
# Inplace means reassign values to df_old or df_new after dropping columns
# axis=1 y axis

# Drop old stars column
df.drop(columns=['time'], axis=1, inplace=True)

# Display data after dropping stars column
df
```

There are duplicates with different release dates, will keep the last release date by appying drop_duplicates() method with the same subset and using keep=last to keep the last release date.

```python
# Drop duplicated rows keeping the last release date
df.drop_duplicates(subset=subset_columns, keep='last', inplace=True)
```
